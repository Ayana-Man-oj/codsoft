```python
# Rule-Based Chatbot

print("Hello! I am your simple chatbot.")
print("Type 'bye' to exit.\n")

while True:
    user_input = input("You: ").lower()

    # Greeting
    if user_input in ["hi", "hello", "hey"]:
        print("Bot: Hello! How can I help you today?")

    # Asking name
    elif "your name" in user_input:
        print("Bot: I am a rule-based chatbot created for Task 1.")

    # Asking about time
    elif "time" in user_input:
        import datetime
        current_time = datetime.datetime.now().strftime("%H:%M:%S")
        print("Bot: Current time is", current_time)

    # Asking about date
    elif "date" in user_input:
        import datetime
        today = datetime.date.today()
        print("Bot: Today's date is", today)

    # Asking about help
    elif "help" in user_input:
        print("Bot: I can respond to greetings, tell time/date, and answer simple questions.")

    # Asking about chatbot
    elif "what is a chatbot" in user_input:
        print("Bot: A chatbot is a program that simulates conversation with users.")

    # Exit condition
    elif user_input == "bye":
        print("Bot: Goodbye! Have a nice day ðŸ˜Š")
        break

    # Default response
    else:
        print("Bot: Sorry, I don't understand that. Please try something else.")
```
